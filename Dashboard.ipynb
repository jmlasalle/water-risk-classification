### Grid of points

```python
# from https://gis.stackexchange.com/questions/269243/creating-polygon-grid-using-geopandas

xmin,ymin,xmax,ymax = data.total_bounds

cols = list(range(int(np.floor(xmin)), int(np.ceil(xmax)+1000), 10000))
rows = list(range(int(np.floor(ymin)), int(np.ceil(ymax)+1000), 10000))
rows.reverse()

points = []

for x in cols:
    for y in rows:
        points.append(Point(x,y))
```

```python
len(points)
```

```python
grid = gpd.GeoDataFrame({'geometry': points})
grid.crs = {'init':'epsg:6933'}
```

```python
grid.sindex
```

```python
grid_joined = gpd.sjoin(grid, data, how="inner", op='intersects')
```

```python
grid_joined.shape
```

```python
mno.bar(grid_joined)
```

```python
grid_joined['x'] = grid_joined.geometry.x
grid_joined['y'] = grid_joined.geometry.y
```

```python
grid_df = pd.DataFrame(grid_joined.drop(columns='geometry'))
grid_df["PR_labels"]=grid_df["PR_labels"].astype("category")
grid_df["deplete_labels"]=grid_df["deplete_labels"].astype("category")
grid_df["floodrought_labels"]=grid_df["floodrought_labels"].astype("category")
```

```python
grid_df.columns
```

```python
grid_df.to_csv("./data/dashboard_data.csv")
```

```python
color_key = {'0':'aqua', '1':'lime',  '2':'red', '3':'fuchsia', '4':'yellow' }

cvs = ds.Canvas(plot_width=400, plot_height=400)
agg = cvs.points(grid_df, 'x', 'y', ds.count_cat('PR_labels'))
img = tf.shade(agg, cmap=color_key, how='linear')
img
```
